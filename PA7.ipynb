```python
import math
class calculator():
    
    def __init__(self,input1,input2):
        self.input1=input1
        self.input2=input2
    def modulo(self):
        return self.input1 % self.input2
    def combination(self):
        return math.comb(round(self.input1),round(self.input2))
    def permutation(self):
        return math.perm(round(self.input1),round(self.input2))
    def absolute_value(self):
        return math.fabs(self.input1),math.fabs(self.input2)
    def actual_rounding(self):
        return round(self.input1),round(self.input2)
    def truncate(self):
        return math.trunc(self.input1),math.trunc(self.input2)
    def round_down(self):
        return math.floor(self.input1),math.floor(self.input2)
    def round_up(self):
        return math.ceil(self.input1),math.ceil(self.input2)
    def integer_squareroot(self):
        return math.isqrt(round(self.input1)),math.isqrt(round(self.input2))
    def factorial(self):
        return math.factorial(round(self.input1)),math.factorial(round(self.input2))
    def exponential(self):
        return math.pow(self.input1,self.input2)
    def squareroot(self):
        return math.sqrt(round(self.input1)),math.sqrt(round(self.input2))
    def addition(self):
        return self.input1+self.input2
    def substraction(self):
        return self.input1-self.input2
    def multiplication(self):
        return self.input1*self.input2
    def division(self):
        return self.input1/self.input2

input_1=float(input("Enter first number: "))
input_2=float(input("Enter second number: "))
my_calc=calculator(input_1,input_2)

choice=1
while choice!=0:
    print()
    print("0. Exit")
    print("1. modulo")
    print("2. combinations")
    print("3. permutations")
    print("4. absolute value")
    print("5. actual rounding")
    print("6. truncate")
    print("7. rounddown")
    print("8. roundup")
    print("9. integer squareroot")
    print("10. factorial")
    print("11. exponential")
    print("12. squareroot")
    print("13. addition")
    print("14. substraction")
    print("15. multiplication")
    print("16. division")
    choice=int(input("Enter your choice...."))
    
    if choice==1:
        print(f"\nResult: {my_calc.modulo()}")
    
    elif choice==2:
        print(f"\nResult: {my_calc.combination()}")
    
    elif choice==3:
        print(f"\nResult: {my_calc.permutation()}")
    
    elif choice==4:
        print(f"\nResult: {my_calc.absolute_value()}")
   
    elif choice==5:
        print(f"\nResult: {my_calc.actual_rounding()}")
    
    elif choice==6:
        print(f"\nResult: {my_calc.truncate()}")
    
    elif choice==7:
        print(f"\nResult: {my_calc.round_down()}")
    
    elif choice==8:
        print(f"\nResult: {my_calc.round_up()}")

    elif choice==9:
        print(f"\nResult: {my_calc.integer_squareroot()}")
    
    elif choice==10:
        print(f"\nResult: {my_calc.factorial()}")
    
    elif choice==11:
        print(f"\nResult: {my_calc.exponential()}")
    
    elif choice==12:
        print(f"\nResult: {my_calc.squareroot()}")
    
    elif choice==13:
        print(f"\nResult: {my_calc.addition()}")
    
    elif choice==14:
        print(f"\nResult: {my_calc.substraction()}")
    
    elif choice==15:
        print(f"\nResult: {my_calc.multiplication()}")
    
    elif choice==16:
        print(f"\nResult: {my_calc.division()}")

    elif choice==0:
        print("Exit")
        
    else:
        print("Sorry. invalid input.")
```
